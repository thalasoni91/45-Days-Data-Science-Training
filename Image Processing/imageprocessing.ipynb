```python
## SCALING: changes the size of an image either enlarging or shrinking
## scaled = cv2.resize(img,(new_width,new_height))

scaled_image = cv2.resize(img,(200,200))
plt.imshow(scaled_image)
```

```python

scaled_image = cv2.resize(img,(0,0),fx=0.5,fy=0.5)
plt.imshow(scaled_image)
```

```python
import numpy as np
rows, cols = img.shape[:2]
tx, ty = 1000,500
M = np.float32([[1,0,tx],[0,1,ty]])

translated_image = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(translated_image)
```

```python
center = (cols//2,rows//2)
angle = 45
scale = 0.5     ##scale -- zoom in zoom out
M = cv2.getRotationMatrix2D(center, angle, scale)

#affine -- translation, rotation, scale, shearing
rotated_image = cv2.warpAffine(img, M, (cols,rows))
plt.imshow(rotated_image)
```

```python
##flipping -- mirror image
flip_image = cv2.flip(img,0)
plt.imshow(flip_image)
```

```python
## cropping an image -- slicing an array
cropped_image = img[100:200,200:300]
plt.imshow(cropped_image)
```

```python
# RGB to color image conversion

img2 = cv2.imread(r"C:\Users\Panks\Downloads\ChatGPT Image Jun 29, 2025, 02_33_48 PM.png")
gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img2)
plt.title("original image")
plt.subplot(1,2,2)
plt.imshow(gray, cmap="gray")
plt.title("gray image")
plt.show()
```

```python
##HISTOGRAMS === distribution of pixel intensity value
color = ("b","g","r")
for i, col in enumerate(color):
    hist = cv2.calcHist([img2],[i],None,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
##erosion -- remove white pixels from boundary of an object
##dilation -- opposite of erosion
```

```python
kernel = np.ones((5,5),np.uint8)
erosion = cv2.erode(img, kernel,iterations = 10)
dilate = cv2.dilate(img,kernel,iterations = 10)
plt.imshow(erosion,cmap="gray")
plt.imshow(dilate,cmap="gray")
```
