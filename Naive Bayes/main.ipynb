```python
# Core Libraries
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import joblib

# Scikit-learn modules
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# NLTK modules
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

nltk.download('stopwords')
```

```python
x,y = [],[]
with open("SMSSpamCollection.txt","r") as file:
    data = file.read()
    #iterate through each line
    for data_item in data.split("\n"):
#it will check if the line contains exactly 2 elements seperated by a tab
        if len(data_item.split("\t")) == 2:
            #append the first element(label) to y
            y.append(data_item.split("\t")[0])
            #append the second element(text) to x
            x.append(data_item.split("\t")[1])
            
print("success")
```

```python
len(x),len(y)
```

```python
#prep your data
data_dict = {"mail_message":x,"label":y}
df = pd.DataFrame(data_dict)
df.head()
```

```python
df.info()
```

```python
for i in range(0,len(df)):

    message = re.sub("[^a-zA-Z]"," ",df["mail_message"][i])  #sub - substitute 
    message = message.lower()
    df["mail_message"][i] = message
df.head()
```

```python
df["label"].value_counts()
```

```python
ham = df[df["label"] == "ham"]
spam = df[df["label"] == "spam"].sample(n=4827)

df = pd.concat([ham,spam],axis=0)
df["label"].value_counts()
```

```python
df["label"] = df["label"].map({"ham":0,"spam":1})
df.head()
```

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
x = df["mail_message"]
y = df["label"]
```

```python
x = x.values
```

```python
x[0:5]
```

```python
Bow_obj = CountVectorizer(
    max_features=2000,
    stop_words=stopwords.words("english")
)

x = Bow_obj.fit_transform(x).toarray()
x.shape
```

```python
joblib.dump(Bow_obj,"vectorizer.lb")
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2
)
```

```python
BNB = BernoulliNB()
BNB.fit(x_train, y_train)
```

```python
BNB.score(x_test, y_test)
```

```python
BNB.score(x_train, y_train)
```

```python
pred = BNB.predict(x_test)
pred
```

```python
compare_dict = {"actual":y_test,"pred":pred}
compare_df = pd.DataFrame(compare_dict)
compare_df.head()
```
