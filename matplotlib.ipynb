```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker = "s",ms = 10,
         mfc = "red",mec = "green",
         color = "black",
         lw = 3,ls = ":",
         alpha = 0.7,label = "lineplot")
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
