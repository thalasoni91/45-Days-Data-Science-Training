```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://github.com/torvalds"
res = req.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
repos = soup.find_all("span", class_="repo")
for repo in repos:
    print(repo.text.strip())
```

```python
url = "https://quotes.toscrape.com/"
res = req.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
quotes = soup.find_all("div", class_="quote")
```

```python
data = []
for quote in quotes:
    text = quote.find("span", class_="text").text
    author = quote.find("small", class_="author").text
    data.append({"text": text, "author": author})
    
df = pd.DataFrame(data)
df.to_csv("quotes.csv", index=False)
```

```python
import time
##this is a fake browser entity
#it will help us to avoid getting blocked by the website
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
```

```python
search_query = "laptops"
url = f"https://www.flipkart.com/search?q={search_query}"
```

```python
res = req.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
# check response status
if res.status_code == 200:
    print("Successfully fetched the page")
```

```python
products = soup.find_all("div", class_="_1AtVbE")
```

```python
data = []
for product in products:
    name = product.find("a", class_="_4rR01T")
    price = product.find("div", class_="_30jeq3")
    if name and price:
        data.append({
            "name": name.text.strip(),
            "price": price.text.strip()
        })
        
df = pd.DataFrame(data)
df.to_csv("laptops.csv", index=False)
```
