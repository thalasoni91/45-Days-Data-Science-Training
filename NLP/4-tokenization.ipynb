```python
## tokenization
# Tokenization is the process of breaking down text into smaller units, such as words or sentences.
# It is a crucial step in natural language processing (NLP) as it allows for the analysis of text data at a granular level.
```

```python
corpus = "Hello, how are you? I am fine. What about you? I am also fine. I am happy to see you. I am happy to see you too. I am happy to see you again. I am happy to see you again and again. I am happy to see you again and again and again. I am happy to see you again and again and again and again. I am happy to see you again and again and again and again and again. I am happy to see you again and again and again and again and again and again. I am happy to see you again and again and again and again and again and again and again. I am happy to see you again and again and again and again and again and again and again and again."
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
doc = sent_tokenize(corpus)
for i in doc:
    print(i)
    print('\n')
    print(word_tokenize(i))
    print('\n\n')
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
print(tokenizer.tokenize(corpus))
```
