```python
#Stemming -- work of this module is to reduce words to their root form
## plays an important role in text processing and natural language understanding.
# plays, playing, played -> play
# running, run, runs -> run
```

```python
words = ["play", "playing", "played", "run", "running", "runs", "happens", "happened", "happening", "history"]

# history here converted to histori whch is not correct
# so we need to use a better stemmer then Porter Stemmer
```

```python
## porter stemmer
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for word in words:
    print(word, "---------->", ps.stem(word))
```

```python
from nltk.stem import RegexpStemmer
rs = RegexpStemmer('ing$|s$|e$|able$|ible$|ed$', min=4)
for word in words:
    print(word, "---------->", rs.stem(word))
```

```python
#updated version of stemmer
# using Snowball Stemmer which is more accurate than Porter Stemmer
# different from porter stemmer as it uses a more sophisticated algorithm
# and is designed to handle a wider range of languages and word forms.

from nltk.stem import SnowballStemmer
ss = SnowballStemmer("english")

for word in words:
    print(word, "---------->", ss.stem(word))
```
