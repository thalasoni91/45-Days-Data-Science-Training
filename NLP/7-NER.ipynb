```python
#name entity recognition
# This is used to identify and classify named entities in text into predefined categories such as person names, organizations, locations, dates, etc.
# NER is a crucial step in many NLP applications, such as information extraction, question answering, and text summarization.
# It helps in understanding the context and meaning of the text by recognizing important entities and their relationships within the text.
```

```python
paragraph = """
Machu Picchu is an ancient Incan citadel located high in the Andes Mountains of Peru. 
Believed to have been built in the 15th century, it remained hidden from the outside world for centuries. 
The site features polished dry-stone walls, terraces, and temples that reflect the architectural genius of the Inca civilization. 
Surrounded by lush greenery and steep peaks, Machu Picchu is a UNESCO World Heritage Site and one of the most iconic tourist destinations in the world. 
Its mysterious origin and breathtaking beauty continue to attract historians, archaeologists, and travelers from around the globe.
"""
```

```python
import nltk
from nltk.tokenize import word_tokenize 
from nltk import pos_tag
from nltk import ne_chunk
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('max_ne_chunker')
nltk.download('words')
import nltk
nltk.download('averaged_perceptron_tagger_eng')

nltk.download('omw-1.4') 
nltk.download('maxent_ne_chunker_tab')

```

```python
words = nltk.word_tokenize(paragraph)
tag_elements = nltk.pos_tag(words)
```

```python
entities = ne_chunk(tag_elements, binary=True)  # Assign to 'entities'

# Print the named entity tree as plain text
print(entities)
```
