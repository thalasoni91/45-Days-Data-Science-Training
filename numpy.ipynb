```python
## indexing and slicing
## 1-D array (similar concept to list)

ar2 = np.arange(1,10)
print(ar2)

#indexing
print(lst[3])
print(lst[-2])

#slicing
print(ar2[1:4])
print(ar2[:-3])
```

```python
## indexing and slicing
## 2-D array  --  [row start : row end : row skip , col start : col end : col skip]

arr11 = np.arange(60).reshape(10,6)
print(arr11[3:4])
print(arr11[3:6, 2:])
print(arr11[3:4, 2:3]) ## since the output is in array we want in int so we do
print(arr11[3:4, 2:3][0][0])
```

```python
arr = np.arange(12).reshape(4, 3)
print(arr)
r = arr[::-1, ::-1]  #rows reversed, columns reversed
print(r)
```
