```python
import copy 
obj = (2,3,4,5,6,7,8,9,10)
```

```python
copy.copy(obj)     # Shallow
copy.deepcopy(obj) # Deep
```

```python
import copy
a = [[1,2],[3,4]]
b = copy.copy(a)
b[0][0] = 99   # affects a too
b

```

```python
c = copy.deepcopy(a)
c[0][0] = 100  # doesn't affect a
c
```

```python
s = "hello"
rev = ""
for ch in s: # ch stands for character
    rev = ch + rev
print(rev)  # olleh
```

```python
lst = [1, 2, 2, 3, 1, 2]
freq = {}

for i in lst:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print(freq)  # {1:2, 2:3, 3:1}
```
